# Bias-Variance tradeoff
#### Bias are the simplifying assumptions made by the model to make the target function easier to them. Variance is the amount that the estimate of the target function will change if different training data was used. The goal of any supervised machine learning algorithm is to achieve low bias and low variance. In turn the algorithm should achieve good prediction performance.

Over fitted model ==> high variance, low bias

Under fitted model ==> low variance, high bias

Best fitted model ==> minimum bias, minimum variance
