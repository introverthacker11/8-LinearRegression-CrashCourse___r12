###
## Importing Necessary things
###

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

###
## Previous Codes to Move On
###

```python
a = pd.read_csv('california500')
b = a.corr()
x = a.drop(['Price'], axis = 1)
x.head()

y = a['Price']
y.head()

x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2, random_state = 0)

x_train.shape, x_test.shape, y_train.shape, y_test.shape

model = LinearRegression()
model.fit( x_train, y_train )

y_predict = model.predict( x_test )
print('DataFrame of orignal and predicted House Prices(y_test) of Test DataSet')
pd.DataFrame({'Orignal y_test Values' : y_test, 'Predicted y_test Values' : model.predict(x_test)})
```

###
## Getting Performance Score (r2) of Our Current Model
###

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_predict)
print(f"Model's Performance Score: {r2 * 100:.3f} %")
```

###
## MAE, MSE & RMSE
###

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
mean_squared_error(y_test, y_predict)
```

```python
np.sqrt(mean_squared_error(y_test, y_predict))
```

###
## Matplotlib Plot for Orignal y_test vs Predicted y_test
###

```python
plt.subplots(figsize = (16,6))
x_points = list(range(len(y_test)))
print(f'\t\t\t\t\tRange of y_test\'s data points:\n\n{x_points}')
print(f'\n\t\t\t\t\tOrignal 20% data points of y:\n\n{y_test}')
print(f'\n\t\t\t\t\tPredicted 20% data points of y:\n\n{y_predict}')
plt.plot(x_points, y_test, label = 'orignal y_test', marker = 'o')
plt.plot(x_points, y_predict, label = 'predicted y_test', marker = 'o')
plt.legend()
plt.grid(True)
```
