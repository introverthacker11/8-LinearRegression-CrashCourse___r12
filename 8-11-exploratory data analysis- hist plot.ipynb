```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
a = pd.read_csv('california500')

# Determine the number of columns
num_columns = len(a.columns)

# Set the number of rows and columns for subplots
rows = (num_columns // 7) + 1 if num_columns % 7 != 0 else num_columns // 7
cols = 7

# Create subplots
fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(16, 6 * rows))
ax = ax.flatten()

# Plot the histograms
for index, column in enumerate(a.columns):
    ax[index].hist(a[column])
    ax[index].set_title(column)

# Hide any unused subplots
for i in range(index + 1, len(ax)):
    fig.delaxes(ax[i])

plt.tight_layout()
plt.show()
```
