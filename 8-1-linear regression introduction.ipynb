# Linear Regression
**Let's first uderstand what exactly regression means. It is a statistical method used in finance, investing, and other diciplines that attempts to determine the strength and character of the relationship between one dependent variable (usually denoted by Y) and a series of other variables known as independent variables.**

**Linear Regression is a statistical technique where based on a set of independent variable(s) a dependent variable is predicted.**

### 
## Linear Regression is Used For:

- **Prediction: Estimating the value of a dependent variable based on the value(s) of one or more independent variables.**
* **Trend Analysis: Identifying and understanding trends and relationships within data.**
- **Forecasting: Predicting future values based on historical data.**
* **Evaluating Relationships: Assessing the strength and direction of relationships between variables.**
- **Optimization: Finding the best combination of independent variables to maximize or minimize a particular outcome.**
* **Inferential Analysis: Making inferences about the population from a sample by understanding the impact of independent variables on a dependent variable.**
